```python
import os
import shutil
import random

# Adjust these paths for your actual Kaggle directory structure
sea_train_dir = '/kaggle/input/seadronessee-sample-dataset-preprocessed/preprocessed_images/train'
hit_train_dir = '/kaggle/input/hituav-a-highaltitude-infrared-thermal-dataset/hit-uav/images/train'
hit_test_dir = '/kaggle/input/hituav-a-highaltitude-infrared-thermal-dataset/hit-uav/images/test'

out_trainA = '/kaggle/working/pytorch-CycleGAN-and-pix2pix/datasets/rgb2ir/trainA'
out_trainB = '/kaggle/working/pytorch-CycleGAN-and-pix2pix/datasets/rgb2ir/trainB'
out_testA = '/kaggle/working/pytorch-CycleGAN-and-pix2pix/datasets/rgb2ir/testA'
out_testB = '/kaggle/working/pytorch-CycleGAN-and-pix2pix/datasets/rgb2ir/testB'

os.makedirs(out_trainA, exist_ok=True)
os.makedirs(out_trainB, exist_ok=True)
os.makedirs(out_testA, exist_ok=True)
os.makedirs(out_testB, exist_ok=True)

# Get RGB image filenames
sea_imgs = os.listdir(sea_train_dir)
random.shuffle(sea_imgs)
trainA_imgs = sea_imgs[:10000]
testA_imgs = sea_imgs[10000:12000]

# Get IR image filenames
hit_train_imgs = os.listdir(hit_train_dir)
random.shuffle(hit_train_imgs)
trainB_imgs = hit_train_imgs[:10000]

hit_test_imgs = os.listdir(hit_test_dir)
random.shuffle(hit_test_imgs)
testB_imgs = hit_test_imgs[:2000]

# Copy files
for f in trainA_imgs:
    shutil.copy(os.path.join(sea_train_dir, f), out_trainA)
for f in testA_imgs:
    shutil.copy(os.path.join(sea_train_dir, f), out_testA)
for f in trainB_imgs:
    shutil.copy(os.path.join(hit_train_dir, f), out_trainB)
for f in testB_imgs:
    shutil.copy(os.path.join(hit_test_dir, f), out_testB)

print("Finished copying all images for CycleGAN dataset structure.")
```
