## Downloading the dataset from kaggle

```python
# The following code will only execute
# successfully when compression is complete

import kagglehub

# Download latest version
path = kagglehub.dataset_download("rameezakther/seadronessee-ir-translated-dataset")

print("Path to dataset files:", path)
```

## Yaml file creation

```python
dataset_yaml = """
# SeaDronesSee IR Dataset for YOLOv8
path: /root/.cache/kagglehub/datasets/rameezakther/seadronessee-ir-translated-dataset/versions/3
train: images/train
val: images/val

names:
  0: person
  1: boat
"""
with open("seadronessee_ir.yaml", "w") as f:
    f.write(dataset_yaml)

print("âœ… Dataset YAML file created: seadronessee_ir.yaml")
```

## Training YoloV8m model

```python
from ultralytics import YOLO
model = YOLO("yolov8m.pt")

results = model.train(
    data="seadronessee_ir.yaml",
    epochs=100,
    imgsz=256,
    batch=32,
    device=0,
    project="runs/train",
    workers=2,
)
```
